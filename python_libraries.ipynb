```python
from random import randint
```

```python
class Die():
  def __init__(self, num_sides = 6):
    self.num_sides = num_sides
  def roll(self):
    return randint(1,self.num_sides)
```

```python
die = Die()
results = []
for roll_num in range(15):
  res = die.roll()
  results += [res]
frequencies = []
for value in range(1, die.num_sides+1):
  freq = results.count(value)
  frequencies += [freq]
```

```python
print(results)
print(frequencies)
```
